## Average age of Times nominee at year of award

```python
times_df["Age when recieving honor"] = times_df["Year"] - times_df["Birth Year"]
times_df["Age when recieving honor"].plot(kind = "bar")
```

```python
times_single_person_df = times_df.dropna(subset = ["Birth Year"])
times_single_person_df.isna().sum()
```

```python
times_single_person_df["Age when recieving honor"].mean()
```

```python
import matplotlib.pyplot as plt
times_fig, times_ax= plt.subplots()
times_single_person_df["Age when recieving honor"].plot(kind = "bar", ax = times_ax)
times_ax.axhline(times_single_person_df["Age when recieving honor"].mean(), color = "red")
```

## Oldest Times nominee still allive today  (2022)

```python
times_alive_df = times_single_person_df[times_single_person_df["Death Year"].isna()]
times_alive_df
```
